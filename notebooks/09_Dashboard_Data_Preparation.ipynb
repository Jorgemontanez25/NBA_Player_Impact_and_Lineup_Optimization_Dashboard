# Dashboard Data Preparation

## Overview

This notebook prepares the final datasets that will be used in the Streamlit dashboard.

The goal is to convert the project’s analytical outputs into clean, lightweight, dashboard-ready files that can be loaded quickly and displayed interactively.

This step separates analysis from application logic, making the final dashboard faster, cleaner, and easier to maintain.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_colwidth", 200)
```

##  Load processed datasets

We load the final processed outputs from previous notebooks. These datasets represent the cleaned player, lineup, and pair-level analysis that will feed the dashboard.

```python
lineup_perf = pd.read_csv("../data/processed/lineup_performance.csv")
reliable_lineups = pd.read_csv("../data/processed/reliable_lineups.csv")
player_impact = pd.read_csv("../data/processed/player_impact.csv")
reliable_player_impact = pd.read_csv("../data/processed/reliable_player_impact.csv")
pair_perf = pd.read_csv("../data/processed/pair_performance.csv")
reliable_pairs = pd.read_csv("../data/processed/reliable_pair_performance.csv")
stable_lineups = pd.read_csv("../data/processed/stable_lineups_modeling.csv")
high_value_lineups = pd.read_csv("../data/processed/high_value_lineups.csv")
underused_watchlist = pd.read_csv("../data/processed/underused_lineups_watchlist.csv")
```

## Create dashboard summary KPIs

This summary table provides high-level counts that can be displayed on the dashboard landing page.

```python
dashboard_summary = pd.DataFrame({
    "metric": [
        "Total lineups analyzed",
        "Reliable lineups",
        "Players analyzed",
        "Reliable players",
        "Player pairs analyzed",
        "Reliable player pairs",
        "High-value lineups",
        "Underused watchlist lineups"
    ],
    "value": [
        len(lineup_perf),
        len(reliable_lineups),
        len(player_impact),
        len(reliable_player_impact),
        len(pair_perf),
        len(reliable_pairs),
        len(high_value_lineups),
        len(underused_watchlist)
    ]
})

dashboard_summary
```

##  Preview dashboard-ready tables

Before saving, we preview the main dashboard datasets to confirm that they contain clean, interpretable fields for the app.

```python
dashboard_players = reliable_player_impact[[
    "player_name",
    "total_net_points",
    "total_duration",
    "total_stints",
    "impact_per_60_sec"
]].copy()

dashboard_lineups = reliable_lineups.copy()

dashboard_pairs = reliable_pairs[[
    "player_1_name",
    "player_2_name",
    "total_net_points",
    "total_duration",
    "total_stints",
    "impact_per_60"
]].copy()

dashboard_players["impact_per_60_sec"] = dashboard_players["impact_per_60_sec"].round(2)
dashboard_lineups["net_per_60"] = dashboard_lineups["net_per_60"].round(2)
dashboard_pairs["impact_per_60"] = dashboard_pairs["impact_per_60"].round(2)

display(dashboard_players.head())
display(dashboard_lineups.head())
display(dashboard_pairs.head())
```

##  Sort tables for dashboard display

The dashboard tables are sorted so the strongest player, lineup, and pair results appear first by default.

```python
dashboard_players = dashboard_players.sort_values(
    "impact_per_60_sec", ascending=False
)

dashboard_lineups = dashboard_lineups.sort_values(
    "net_per_60", ascending=False
)

dashboard_pairs = dashboard_pairs.sort_values(
    ["impact_per_60", "total_duration"],
    ascending=[False, False]
)
```

## Save dashboard-ready datasets

The final dashboard files are saved in the processed data folder. These files are designed to be loaded directly by the Streamlit app without additional transformation.

```python
dashboard_summary.to_csv("../data/processed/dashboard_summary.csv", index=False)
dashboard_players.to_csv("../data/processed/dashboard_players.csv", index=False)
dashboard_lineups.to_csv("../data/processed/dashboard_lineups.csv", index=False)
dashboard_pairs.to_csv("../data/processed/dashboard_pairs.csv", index=False)
high_value_lineups.to_csv("../data/processed/dashboard_high_value_lineups.csv", index=False)
underused_watchlist.to_csv("../data/processed/dashboard_underused_lineups.csv", index=False)

print("Dashboard datasets created successfully ✔")
```
